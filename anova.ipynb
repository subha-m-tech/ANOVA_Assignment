```python
from scipy import stats
```

Statement:
Ho  - There is no significant difference between ssc_p, ssc_p and degree_p
H1  - There is a significant difference between ssc_p, ssc_p and degree_p
Accept/Reject: if p < 0.05 then reject Ho (null hypothesis), or else accept Ho (null hypothesis) and reject H1 (alternate hypothesis)

```python
stats.f_oneway(dataset["ssc_p"], dataset["hsc_p"], dataset["degree_p"])
```

Here p value is 0.51 > 0.05 so we can accept Ho

# Assignment  for two-way classification
# Ho - There is no significant difference between ssc_p, ssc_p and degree_p columns and also Salary

 # H1 - There is a significant difference between ssc_p, ssc_p and degree_p and also Salary

# Accept/Reject: if p < 0.05 then reject Ho (null hypothesis), or else accept Ho (null hypothesis) and reject H1 (alternate hypothesis)

```python
# OLS ( Ordinary Least Squares) aims to find the line (or hyperplane in higher dimensions) that best fits 
# the data points by minimizing the sum of the squared distances 
# between the observed values and the values predicted by the model
# formula is a string expression that defines the relationship between the dependent variable and the independent variables in your model
from statsmodels.formula.api import ols
model = ols(formula= 'salary ~ ssc_p + hsc_p + degree_p', data=dataset[["ssc_p", "hsc_p", "degree_p","salary"]]).fit()  
```

```python
model
```

```python
import statsmodels.api as sm
```

```python
sm.stats.anova_lm(model, typ=3)
```

# Here p value of all the columns > 0.05 so we can accept Ho
